```python
weatherData.dropna(axis=0, how='all', thresh=None, subset=None, inplace=True)
```

```python
weatherData.dropna(axis=1, how='all', thresh=None, subset=None, inplace=True)
```

```python
weatherData.drop_duplicates()
```

```python
cleanWeatherData = weatherData.iloc[:160000, :]
```

```python
cleanWeatherData.shape
```

```python
cleanWeatherData.to_csv("cleanWeatherData.csv")
```

```python
stationData = cleanWeatherData[['wsid', 'wsnm', 'inme' ]]
stationData = stationData.drop_duplicates()
# locationData.to_csv("snowFlakeSchema/locData.csv")
stationData
```

```python
locationData = cleanWeatherData[['wsid', 'elvt', 'lat','lon','city','prov']]
locationData = locationData.drop_duplicates()
locationData.insert(0, 'locId', range(880, 880 + len(locationData)))
```

```python
locationData
```

```python
hourlyData = cleanWeatherData[['mdct', 'date','hr']]
hourlyData = hourlyData.drop_duplicates()
hourlyData
```

```python
dailyData = cleanWeatherData[['date','yr','mo','da' ]]
dailyData = dailyData.drop_duplicates()
dailyData
```

```python
windQuality = cleanWeatherData[['wdsp','wdct','gust' ]]
windQuality.insert(0, 'windId', range(70, 70 + len(windQuality)))
windQuality
```

```python
avgHourlyTempData = cleanWeatherData[['mdct','temp','gbrd','stp','dewp','hmdy' ]]
avgHourlyTempData = avgHourlyTempData.drop_duplicates()
avgHourlyTempData
```

```python
maxHourlyTempData = cleanWeatherData[['mdct','smax','tmax','dmax','hmax' ]]
maxHourlyTempData = maxHourlyTempData.drop_duplicates()
maxHourlyTempData
```

```python
minHourlyTempData = cleanWeatherData[['mdct','smin','tmin','dmin','hmin' ]]
minHourlyTempData = minHourlyTempData.drop_duplicates()
minHourlyTempData
```

```python
hourlyWeatherData = cleanWeatherData[['mdct','wsid','prcp' ]]
hourlyWeatherData.insert(0, 'windId', range(70, 70 + len(windQuality)))
hourlyWeatherData
```

```python
stationData.to_csv("snowFlakeSchema/stationData.csv", index = False)
locationData.to_csv("snowFlakeSchema/locationData.csv", index = False)
hourlyData.to_csv("snowFlakeSchema/hourlyData.csv", index = False)
dailyData.to_csv("snowFlakeSchema/dailyData.csv", index = False)
windQuality.to_csv("snowFlakeSchema/windQuality.csv", index = False)
avgHourlyTempData.to_csv("snowFlakeSchema/avgHourlyTempData.csv", index = False)
maxHourlyTempData.to_csv("snowFlakeSchema/maxHourlyTempData.csv", index = False)
minHourlyTempData.to_csv("snowFlakeSchema/minHourlyTempData.csv", index = False)
hourlyWeatherData.to_csv("snowFlakeSchema/hourlyWeatherData.csv", index = False)
```
